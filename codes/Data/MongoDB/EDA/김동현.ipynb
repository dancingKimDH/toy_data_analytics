```python
import re
df_D20_hyper.loc[:, 'BTH_YYYY'] = df_D20_hyper['BTH_YYYY'].astype(str)
pattern = r'^(\w{4})'
df_D20_hyper['BTH_YYYY'] = df_D20_hyper['BTH_YYYY'].str.extract(pattern)
df_D20_hyper['BTH_Year'] = df_D20_hyper['BTH_YYYY']
df_D20_hyper['BTH_Year']
df_D20_hyper['BTH_Year'] = pd.to_datetime(df_D20_hyper['BTH_Year'])
# df_D20_hyper['DTH_YYYYMM']
def age(year):
    if year.year >= 1983:
        return 'People under 40'
    elif 1982 >= year.year >= 1959:
        return 'People in their 40s to 64'
    else:
        return 'People 65 and older'
df_D20_hyper['BTH_Year'] = df_D20_hyper['BTH_Year'].apply(age)
df_D20_hyper[['BTH_Year', 'BTH_YYYY', 'RN_INDI']]
```
